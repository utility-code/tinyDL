```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sh(arr):
    plt.cla()
    plt.clf()
    plt.imshow(arr)
```

```python
def conv2d(img, kernel, stride = 1, padding = 0):
    kernel = np.flipud(np.fliplr(kernel)) # cross correlation
    kernel_x,kernel_y = kernel.shape
    img_x, img_y = img.shape
    
    x_out = int(((img_x - kernel_x +2*padding)/stride)+1)
    y_out = int(((img_y - kernel_y +2*padding)/stride)+1)
    output = np.zeros((x_out, y_out))
    
    if padding !=0:
        padded = np.zeros((img.shape[0]+padding*2 , image.shape[1]+padding*2))
        padded[int(padding):int(-1*padding):int(-1*padding)] = img
    else:
        padded = img
    
    for y in range(img.shape[1]):
        if y>img.shape[1] - kernel_y:
            break
        if y%stride == 0:
            for x in range(img.shape[0]):
                if x>img.shape[0] - kernel_x:
                    break
                try:
                    if x%stride ==0:
                        output[x,y] = (kernel*padded[x:x+kernel_x, y:y+kernel_y]).sum()
                except:
                    break
    return output
```

```python
img = np.random.random((256,256))
kernel = np.ones((3,3))
```

```python
sh(img)
```

```python
%time
sh(conv2d(img, kernel))
```
